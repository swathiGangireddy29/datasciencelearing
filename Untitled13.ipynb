```python
import pandas as pd
import numpy as np
```

```python
df = df = pd.read_csv("netflix_titles.csv")
```

```python
df
```

#part A

```python
df['director'] = df['director'].fillna('Unknown')
df
```

```python
df['cast'] = df['cast'].fillna('Unknown')
df
```

```python
df['country'] = df['country'].fillna('Unknown')
df
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df
```

```python
df['description'] = df['description'].fillna('No Description')
df['date_added'] = df['date_added'].fillna('January 1, 2020')
df
```

#part B

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed', dayfirst=False)
df['AddedYear'] = df['date_added'].dt.year
df
```

```python
df['AddedMonth'] = df['date_added'].dt.month
df
```

```python
df['AddedDay'] = df['date_added'].dt.day
df
```

```python
df['AddedWeekDay'] = df['date_added'].dt.day_name()
df
```

```python
df['AddedQuarter'] = df['date_added'].dt.quarter
df
```
