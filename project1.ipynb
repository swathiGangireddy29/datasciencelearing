```python
import pickle
import gradio as gr
model = pickle.load(open(
    "/content/weighttheightpredmodel.pkl",
    "rb"))

scaler = pickle.load(open(
    "/content/scaler.pkl",
    "rb"
))
```

```python
def predict_height(weight):
  scaled = scaler.transform([[weight]])
  prediction = model.predict(scaled)
  return prediction[0]
app = gr.Interface(
    fn = predict_height,
    inputs = gr.Number(label="enter weight"),
    outputs = gr.Number(label = "predicted height"),
    title = "Weight Height Prediction"

)
app.launch()
```
