```python
#converting data column
df["Date"] = pd.to_datetime(df["Date"])
df.info()
```

```python
#sort the values
dfd = df.sort_values("Date")
dfd
```

```python
#sort date as index
dfd.set_index("Date",inplace = True)
dfd
```

```python
#so we can create missing dates too
dfd = dfd.asfreq("D")
dfd
# code-meaning
# d - Daily
# w - weekly
# m - monthly
# y - yearly
# h - hourly
```

```python
#handing missing values
# day1 = 100
# day2 =
# day3 =
# day4 =160
# 60/3 = 20
#interpolation means filling missing values by using
#the values before and after them and estimating
#what the missing values should be
dfd["Stock_Price"] = dfd["Stock_Price"].interpolate()
dfd["Volume"] = dfd["Volume"].interpolate()
dfd
```

```python
#forwardfill replaces a missing value with previous avilable values
#(ffill)
#backwardpill replaces a missing value with next available valuebfill
```

```python
#convert date features
dfd["Year"] = dfd.index.year
dfd["Month"] = dfd.index.month
dfd["Day"] = dfd.index.day
dfd
dfd["dayofweek"] = dfd.index.dayofweek
dfd
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["weekend"] = (df["Date"].dt.dayofweek>=5).astype(int)
dfd
```

```python
#lag feature
#a lag ferture means use past values to predict future values
dfd["lag1"] = dfd["Stock_Price"].shift(1)
dfd
dfd["lag2"] = dfd["Stock_Price"].shift(2)
dfd
```
