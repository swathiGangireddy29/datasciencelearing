```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = sns.load_dataset('diamonds')
df
```

```python
df.head(5)
df.shape
df.columns
df[['cut', 'color', 'clarity']].dtypes
```

```python
encoder_cut = OneHotEncoder(sparse_output=False)

cut_encoded = encoder_cut.fit_transform(df[['cut']])

cut_df = pd.DataFrame(
    cut_encoded,
    columns=encoder_cut.get_feature_names_out(['cut'])
)
cut_df
```

```python
encoder_color = OneHotEncoder(sparse_output=False)

color_encoded = encoder_color.fit_transform(df[['color']])

color_df = pd.DataFrame(
    color_encoded,
    columns=encoder_color.get_feature_names_out(['color'])
)
color_df
```

```python
encoder_clarity = OneHotEncoder(sparse_output=False)

clarity_encoded = encoder_clarity.fit_transform(df[['clarity']])

clarity_df = pd.DataFrame(
    clarity_encoded,
    columns=encoder_clarity.get_feature_names_out(['clarity'])
)
clarity_df
```

```python
df_numeric = df.drop(columns=['cut', 'color', 'clarity'])
final_df = pd.concat([df_numeric, cut_df, color_df, clarity_df], axis=1)
final_df.head()
final_df.shape
final_df.columns
```

```python
df_remaining = df.drop(columns=['cut', 'color', 'clarity'])
final_df = pd.concat(
    [df_remaining, cut_df, color_df, clarity_df],
    axis=1
)
final_df.head()
final_df.shape
final_df.columns
```
