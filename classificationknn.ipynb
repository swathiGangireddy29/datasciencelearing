```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

```

```python
df = pd.read_csv('/content/knnclassification.csv')
```

```python
X = df.drop('Loan_Status', axis=1)
X
Y = df['Loan_Status']
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state = 42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
```

```python
confusion_matrixh = confusion_matrix(Y_test, Y_pred)
print(confusion_matrixh)
accuracy = accuracy_score(Y_test, Y_pred)
print(accuracy)
precision = precision_score(Y_test, Y_pred,pos_label = "Approved")
print(precision)
recall = recall_score(Y_test, Y_pred,pos_label = "Approved")
print(recall)
f1 = f1_score(Y_test, Y_pred,pos_label = "Approved")
print(f1)
```
