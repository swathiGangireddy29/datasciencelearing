```python
#anova = analysis of variance
#anova helps us compare multpile groups and check whether their
#average values are significantly different from other
from scipy.stats import f_oneway
method_a = [70,72,68,75,75]
method_b = [80,82,78,85,81]
method_c = [90,92,88,95,91]
f_stat,p_value = f_oneway(
    method_a,
    method_b,
    method_c,
)
#fsta tells us whether the gropus are significantly different from each other
print(p_value)
if p_value<0.05:
  print("reject h0")
else:
  print("fail to reject h0")
```

```python
from scipy.stats import chisquare
observed = [20,10,5,5]
expected = [10,10,10,10]
chi2,p_value = chisquare(observed,expected)
print(chi2)
if p_value<0.05:
  print("reject h0")
else:
  print("fail to reject")
```
