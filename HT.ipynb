```python
df["Name"] = df["Name"].fillna("Unknown")
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
df["Qualification"] = df["Qualification"].fillna(df["Qualification"].mode()[0])
df["Experience_Years"] = df["Experience_Years"].fillna(df["Experience_Years"].median())
df["Expected_Salary"] = df["Expected_Salary"].fillna(df["Expected_Salary"].median())
df["City"] = df["City"].fillna(df["City"].mode()[0])
```
