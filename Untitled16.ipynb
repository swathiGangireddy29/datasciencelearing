```python
days = [1,2,3,4,5]
sales = [100,200,300,400,500]
plt.plot(
    days,
    sales,
    color = 'red',
    linestyle = '--',
    marker = 'o',
    markersize = 10,
    linewidth = 3,
    alpha = 0.5,
    label = 'sales'
)
plt.legend()
plt.grid(linestyle = '-', color = 'red' , linewidth=3)
plt.show
```

```python
import pandas as pd
data = {
    "Product":["laptop","Mobile","Tablet","Monitor","Printer"],
    "Sales_2024":[250,300,350,400,450],
    "Sales_2025":[300,350,400,450,500]
}
df = pd.DataFrame(data)
df
```

```python
plt.bar(
    df['Product'],
    df['Sales_2024']
)
plt.show()
```

```python
plt.bar(
    df['Product'],
    df['Sales_2024'],
    color = 'red'
)
plt.show()
```
